- Visualize simulated root length density in both scenarios. The simulated data are from the files:
    * for control: springbarley_rootsim.py
    * for deep loosening: springbarley_rootsimDL.py

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# load simulated rld in control scenario

sim_ctrl = pd.read_csv('sim_data_ctrl.txt', header=None)
```

```python
# load simulated rld in DL scenario

df = pd.read_csv('sim_data_DL.txt', header=None)
df_DL = df.iloc[:,30:50]
```

```python
# rld during shooting in control scenario

DAS_45 = sim_ctrl.iloc[:20,]
DAS_46 = sim_ctrl.iloc[20:40,].reset_index(drop=True)

# rld during flowering in control scenario

DAS_71 = sim_ctrl.iloc[40:60,].reset_index(drop=True)
DAS_73 = sim_ctrl.iloc[60:80,].reset_index(drop=True)
DAS_74 = sim_ctrl.iloc[80:,].reset_index(drop=True)
```

```python
# rld during shooting in DL scenario

DAS45_DL = df_DL.iloc[:20,]
DAS46_DL = df_DL.iloc[20:40,].reset_index(drop=True)

# rld during flowering in DL scenario

DAS71_DL = df_DL.iloc[40:60,].reset_index(drop=True)
DAS73_DL = df_DL.iloc[60:80,].reset_index(drop=True)
DAS74_DL = df_DL.iloc[80:,].reset_index(drop=True)
```

```python
# contour lines in control scenario

x = np.linspace(-47.5, 47.5, 20)
y = np.linspace(-2.5, -97.5, 20)
levels = np.linspace(0, 4, 21)
X, Y = np.meshgrid(x, y)
```

```python
# plots of RLD in control scenario

# during shooting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))
fig.suptitle("RLD During Plant Shooting", fontsize=16, y=1.07)

fig12 = ax1.contourf(X, Y, DAS_45, cmap='YlGnBu', levels=levels)
ax1.set_xlabel("Distance of the row [cm]", fontsize=15, labelpad=15)
ax1.set_ylabel("Depth [cm]", fontsize=15)
ax1.xaxis.tick_top()
ax1.xaxis.set_label_position("top")
cbar = fig.colorbar(fig12, ax=ax1)

fig13 = ax2.contourf(X, Y, DAS_46, cmap='YlGnBu', levels=levels)
ax2.set_xlabel("Distance of the row [cm]", fontsize=15, labelpad=15)
ax2.xaxis.tick_top()
ax2.xaxis.set_label_position("top")
cbar = fig.colorbar(fig13, ax=ax2)
cbar.ax.set_ylabel("Root Length Density [cm cm$^-$$^3$]", fontsize=15, labelpad=15)

plt.savefig("RLD during shooting control.png")
plt.show()
```

```python
# plots of RLD in control scenario

# during flowering
fig2, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5))
fig2.suptitle("RLD During Plant Flowering", fontsize=16, y=1.07)

fig21 = ax1.contourf(X, Y, DAS_71, cmap='YlGnBu', levels=levels)
ax1.set_xlabel("Distance of the row [cm]", fontsize=15, labelpad=15)
ax1.set_ylabel("Depth [cm]", fontsize=15)
ax1.xaxis.tick_top()
ax1.xaxis.set_label_position("top")
cbar1 = fig2.colorbar(fig21, ax=ax1)

fig22 = ax2.contourf(X, Y, DAS_73, cmap='YlGnBu', levels=levels)
ax2.set_xlabel("Distance of the row [cm]", fontsize=15, labelpad=15)
ax2.xaxis.tick_top()
ax2.xaxis.set_label_position("top")
cbar1 = fig2.colorbar(fig22, ax=ax2)

fig23 = ax3.contourf(X, Y, DAS_74, cmap='YlGnBu', levels=levels)
ax3.set_xlabel("Distance of the row [cm]", fontsize=15, labelpad=15)
ax3.xaxis.tick_top()
ax3.xaxis.set_label_position("top")
cbar1 = fig2.colorbar(fig23, ax=ax3)

cbar1.ax.set_ylabel("Root Length Density [cm cm$^-$$^3$]", fontsize=15, labelpad=15)

plt.savefig("RLD during flowering control.png")
plt.show()
```

```python
# contour lines in DL scenario
x = np.linspace(0, 95, 20)
y = np.linspace(-2.5, -97.5, 20)
levels = np.linspace(0, 4, 21)
X, Y = np.meshgrid(x, y)
```

```python
# plots of RLD in control scenario

# during shooting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))
fig.suptitle("RLD During Plant Shooting", fontsize=16, y=1.07)

fig12 = ax1.contourf(X, Y, DAS45_DL, cmap='YlGnBu', levels=levels)
ax1.set_xlabel("Distance of the row [cm]", fontsize=15, labelpad=15)
ax1.set_ylabel("Depth [cm]", fontsize=15)
ax1.xaxis.tick_top()
ax1.xaxis.set_label_position("top")
cbar = fig.colorbar(fig12, ax=ax1)

fig13 = ax2.contourf(X, Y, DAS46_DL, cmap='YlGnBu', levels=levels)
ax2.set_xlabel("Distance of the row [cm]", fontsize=15, labelpad=15)
ax2.xaxis.tick_top()
ax2.xaxis.set_label_position("top")
cbar = fig.colorbar(fig13, ax=ax2)
cbar.ax.set_ylabel("Root Length Density [cm cm$^-$$^3$]", fontsize=15, labelpad=15)

plt.savefig("RLD during shooting deep loosening.png")
plt.show()
```
